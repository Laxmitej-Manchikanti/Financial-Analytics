```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/Financial Analytics data.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.describe()
```

```python
top_companies = data.sort_values(by='Mar Cap - Crore', ascending=False).head(10)
```

```python
top_companies
```

```python
plt.figure(figsize=(10, 6))
plt.barh(top_companies['Name'], top_companies['Mar Cap - Crore'], color='skyblue')
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Company')
plt.title('Top 10 Companies by Market Capitalization')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest value at the top
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Mar Cap - Crore'], data['Sales Qtr - Crore'], color='blue', alpha=0.6)
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Quarterly Sales (Crore)')
plt.title('Market Capitalization vs. Quarterly Sales')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.barh(top_companies['Name'], top_companies['Sales Qtr - Crore'], color='skyblue')
plt.xlabel('Sales Quarterly (Crore)')
plt.ylabel('Company')
plt.title('Top 10 Companies by Quarterly Sales')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest value at the top
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.pie(top_companies['Mar Cap - Crore'], labels=top_companies['Name'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proportion of Market Capitalization Among Top 10 Companies')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(loc=4)
plt.show()
```
